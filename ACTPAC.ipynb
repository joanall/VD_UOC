```python
temp = df_reshaped.copy()
#temp2 = temp.loc[temp.index.get_level_values("Area") == "Spain"]
temp3 = temp.unstack().to_frame()
temp3 = temp3.reset_index()
temp4 = temp3[temp3['Element'].isin(["Production", "Import Quantity", "Export Quantity"])]
regions = [
    'Australia and New Zealand',
    'Caribbean',
    'Central America',
    'Central Asia',
    'Eastern Africa',
    'Eastern Asia',
    'Eastern Europe',
    'Melanesia',
    'Micronesia',
    'Middle Africa',
    'Northern Africa',
    'Northern America',
    'Northern Europe',
    'Polynesia',
    'South America',
    'South-eastern Asia',
    'Southern Africa',
    'Southern Asia',
    'Southern Europe',
    'Western Africa',
    'Western Asia',
    'Western Europe',
    'Africa',
    'Americas',
    'Asia',
    'Europe',
    'Oceania',
    'World'
]

temp5 = temp4[temp4['Area'].isin(regions)]
# Agrupar per any i sumar els valors del subministrament alimentari
grouped_df = temp5.groupby(['Area','Year',"Element"])[0].sum().reset_index()
grouped_df = grouped_df[grouped_df["Year"] == "2020"]
transposed_df = grouped_df.pivot(index=['Area', 'Year'], columns='Element', values=0).reset_index()
transposed_df
transposed_df.to_csv('pro_imp_exp.csv', index=False)
```
